# Min-Max Scalling

```python
#(value - min)/(max - min)
```

```python
import seaborn as sns 
```

```python
movies_data_temp= movies_data.copy()
```

```python
sns.distplot(movies_data['RATING'])
```

```python
movies_data_temp["RATING"]=(movies_data_temp["RATING"]-movies_data_temp["RATING"].min())/(movies_data_temp["RATING"].max()-movies_data_temp["RATING"].min())
```

```python
sns.distplot(movies_data_temp["RATING"])
```

# Z-score method

```python
#scaled_value= value-mean / std
```

```python
sns.distplot(movies_data['RunTime'])
```

```python
scaled_data=movies_data.copy()
```

```python
##applying the formola
for col in ['RunTime','RATING']:
    scaled_data[col]= (scaled_data[col]-scaled_data[col].mean())/scaled_data[col].std()
    
```

```python
sns.distplot(scaled_data['RATING'])
```
